```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

```python
data = pd.read_csv("3.12. Example.csv")
```

```python
data
```

```python
plt.scatter(data["Satisfaction"], data["Loyalty"])
plt.xlabel("Satisfaction", size=15)
plt.ylabel("Loyalty", size=15)
plt.show()
```

```python
x = data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters["Clusters"] = kmeans.fit_predict(x)
```

```python
plt.scatter(data["Satisfaction"], data["Loyalty"], c=clusters["Clusters"], cmap="rainbow")
plt.xlabel("Satisfaction", size=15)
plt.ylabel("Loyalty", size=15)
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
```

```python
x_scaled
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)

wcss
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(4)
new_clusters = x.copy()
new_clusters["Clusters"] = kmeans.fit_predict(x_scaled)
```

```python
new_clusters.head()
```

```python
plt.scatter(new_clusters["Satisfaction"], new_clusters["Loyalty"], c=new_clusters["Clusters"], cmap="rainbow")
plt.xlabel("Satisfaction", size=15)
plt.ylabel("Loyalty", size=15)
plt.show()
```
