```python
reason_columns = pd.get_dummies(df["Reason for Absence"], drop_first=True)
```

```python
reason_columns["check"] = reason_columns.sum(axis=1)
```

```python
reason_columns.drop(["check"], axis=1, inplace=True)
```

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
df.drop(["Reason for Absence"], axis=1, inplace=True)
```

```python
df.head(10)
```

```python
reason_type1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
df = pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis=1)
```

```python
df
```

```python
df.columns.values
```

```python
columns_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', "Reason_1", "Reason_2", "Reason_3", "Reason_4"]
```

```python
df.columns = columns_names
```

```python
df.head()
```

```python
df_reason_mod = df.copy()
```

### Date

```python
type(df_reason_mod["Date"][0])
```

```python
df_reason_mod["Date"] = pd.to_datetime(df_reason_mod["Date"], dayfirst=True)
```

```python
df_reason_mod["Date"]
```

```python
df_reason_mod.head(10)
```

```python
df_reason_mod.info()
```

```python
month_values = []
for date in df_reason_mod["Date"]:
    month_values.append(date.month)
```

```python
df_reason_mod["Month Values"] = month_values
```

```python
df_reason_mod
```

```python
weekday_values = []
for date in df_reason_mod["Date"]:
    weekday_values.append(date.weekday())
```

```python
df_reason_mod["Day of the Week"] = weekday_values
```

```python
df_reason_mod["Education"].value_counts()
```

```python
df_reason_mod["Education"] = df_reason_mod["Education"].map({1:0,2:1,3:1,4:1})
```

```python
df_reason_mod
```

```python
df_reason_mod["Education"].value_counts()
```
