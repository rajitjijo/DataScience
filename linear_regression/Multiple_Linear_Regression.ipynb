```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
sns.set()
```

```python
data = pd.read_csv("1.02. Multiple linear regression.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
y = data["GPA"]
x1 = data[["SAT", "Rand 1,2,3"]]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

By adding a new variable __Rand 1,2,3__ we have increased the r squared value which seems the model's explanatory power but because our adjusted r squared has gone down that means we have added a value that is does not give us more information hence we have overall reduced the accurcay of the model.

Furthermore, the p value for the rand 1,2,3 is 0.762 and any value above 0.05 is statistically insignificant
