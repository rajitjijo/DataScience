We will create a linear regression which predicts the GPA of the student based on the SAT score

Define the dependent and independent variables

```python
y = data["GPA"]
x1 = data["SAT"]
```

```python
plt.scatter(x1,y)
plt.xlabel("SAT", fontsize=18)
plt.ylabel("GPA", fontsize=18)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.00178*x1 + 0.275
fig = plt.plot(x1, yhat, lw=4, c="orange", label="regression line")
plt.xlabel("SAT", fontsize=18)
plt.ylabel("GPA", fontsize=18)
plt.show()
```

# Using SKlearn

```python
data = pd.read_csv("1.01. Simple linear regression.csv")
```

```python
x = data['SAT'] #features
y = data['GPA'] #targets
```

```python
x.shape, y.shape
```

```python
x = x.values.reshape(-1,1)
y = y.values.reshape(-1,1)
```

```python
x.shape, y.shape
```

### Creating the Linear Regression Model

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

#### RSquared

```python
reg.score(x,y)
```

### Coeffecients

```python
reg.coef_
```

### Intercept

```python
reg.intercept_
```

### Making Predictions

```python
reg.predict([[1740]])
```

```python
multi_data = pd.read_csv("1.02. Multiple linear regression.csv")
```

```python
multi_data.head()
```

```python
y, x = multi_data["GPA"], multi_data[["SAT", "Rand 1,2,3"]]
multi = LinearRegression()
multi.fit(x,y)
```

```python
multi.coef_
```

```python
multi.intercept_
```

### Calculating the R-Squared

#### Formula for adjusted R-Squared
$R^2_{adj} = 1-(1-R^2)*\frac{n-1}{n-p-1}$

__n__ stands for the number of observations <br>
__p__ stands for the number of predictors

```python
r2 = multi.score(x,y)
r2
```

Now using the r squared value we got from the above formula we can calcualte the adjusted r squared value

## Feature Selection

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

If we look above we geet two outputs in index 0 is the fstatistic for the two features and at index 1 is the pvalue for the featrues, any value greater than 0.05 we ignore that feature

```python
p_value = f_regression(x,y)[1]
```
